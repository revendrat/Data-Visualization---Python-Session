```python
# import libraries

import pandas as pd
from pandas_datareader import data as pdr
import matplotlib.pyplot as plt

import yfinance as yf

# Mandatory, override the yf with pdf_override()
yf.pdr_override() 
```

```python
# download Apple stock data as pandas dataframe
df_apple = pdr.get_data_yahoo("AAPL", start="2023-05-01", end="2023-06-09")
```

```python
type(df_apple)
```

```python
df_apple.head()
```

```python
# Plotting line chart
df_apple.plot(y='Close', kind='line', title='Apple Stock Close Price')
```

```python
# Plotting line chart
df_apple.plot(y=['Open','Adj Close'], kind='line', title='Apple Stock Open & Adj-Close Price')
```

```python
# Plotting line chart
df_apple.plot(y=['Open','High', 'Low', 'Adj Close'], kind='line', title='Apple Stock Open High Low Adj-Close Price')
```

```python
# Plotting bar chart
df_apple.plot( y='Volume', kind='bar', title='Apple Stock Volume')
```

```python
# Plotting histogram
df_apple['Close'].plot(kind='hist', title='Apple Close Price Distribution')
```

```python
# Plotting scatter plot
df_apple.plot(x='Close', y='Volume', kind='scatter', title='Apple Close Price vs Volume')
```

```python
# Plotting box plot
df_apple.plot(y='Close', kind='box', title='Apple Close Price Box Plot')
```

```python
# Plotting area chart
df_apple.plot( y='Close', kind='area', title='Apple Close Price Trend: Area Chart')
```

## Read and plot multiple stocks
* Apple
* Microsoft
* Google

```python
# download Apple stock data as pandas dataframe
df_stocks = pdr.get_data_yahoo("AAPL MSFT GOOG", start="2023-05-01", end="2023-06-09")
df_stocks.head()
```

```python
# Plotting line chart
df_stocks.plot(y='Close', kind='line', title='Stocks Close Price')
```

```python
# Plotting bar chart
df_stocks.plot( y='Volume', kind='bar', title='Stock Volume')
```

```python
# Plotting histogram
df_stocks['Close'].plot(kind='hist', title='Close Price Distribution')
```

```python
# Plotting scatter plot
df_stocks.plot(x='Close', y='Volume', kind='scatter', title='Close Price vs Volume')
```

```python
# Plotting box plot
df_stocks.plot(y='Close', kind='box', title='Close Price Box Plot')
```

```python
# Plotting area chart
df_stocks.plot( y='Close', kind='area', subplots=True, title='Close Price Trend: Area Chart')
```

```python
# Plotting line chart
df_stocks.plot(y=['Open','Adj Close'], kind='line', title='Stocks Close Price')
```
